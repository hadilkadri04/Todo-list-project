```python
from PyQt5 import QtWidgets, uic

# Création de l'application
App = QtWidgets.QApplication([])

# Chargement du fichier .ui
Fen = uic.loadUi('Bienvenu.ui')

# fonction qui prend un nom en paramètre
def Bonjour(name):
    Fen.lblmsg.setText(f'Bonjour {name}')

# fonction appelée par le bouton (sans paramètre)
def Appelbnj():
    name = Fen.lName.text()
    Bonjour(name)   # <-- correction ici !

# connecter le bouton
Fen.btnOk.clicked.connect(Appelbnj)

# Affichage UI
Fen.show()

# Exécution de l'application
App.exec_()
```
